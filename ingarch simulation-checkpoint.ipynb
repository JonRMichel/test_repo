```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega, alpha , beta =  0.67 ,0.03 ,0.97
```

```python
simulation_df = pd.DataFrame(columns = ['y','lam'])
```

```python
simulation_df.loc[0,'lam'] = omega
simulation_df.loc[0,'y'] = np.random.poisson(omega,1)[0]

for time_index in range(1,100):
    lam = omega + alpha * simulation_df.loc[time_index -1 , 'y'] + beta*simulation_df.loc[time_index-1,'lam']
    
    simulation_df.loc[time_index,'lam'] = lam
    simulation_df.loc[time_index,'y'] = np.random.poisson(lam,1)[0]
    
plt.rcParams.update({'font.size': 22})
fig, ax = plt.subplots(figsize = (30,10))
ax.set_title('Simulation with (omega,alpha,beta) = (0.67 ,0.03 ,0.97)')
simulation_df['y'].plot(ax = ax)
```
